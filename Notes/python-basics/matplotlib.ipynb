```python
import matplotlib.pyplot as plt
import numpy as np

# 生成从 0 到 2 的 500 个等间距 y 值
y = np.linspace(0, 2, 500)

# 定义两条曲线：x1 = sin(2πy), x2 = 1.1 * sin(3πy)
x1 = np.sin(2 * np.pi * y)
x2 = 1.1 * np.sin(3 * np.pi * y)

# 创建图形对象
fig = plt.figure()

# 添加子图（1行1列的第1个子图）
ax = fig.add_subplot(111)

# 绘制第一条曲线（x1, y），黑色、线宽1、实线
ax.plot(x1, y, color="k", lw=1, ls="-")

# 绘制第二条曲线（x2, y），同样风格
ax.plot(x2, y, color="k", lw=1, ls="-")

# 用 fill_betweenx 在 x1 <= x2 的区域之间填充蓝色
# 参数说明：
# - y: 纵轴坐标
# - x1, x2: 对应于每个 y 的两条横坐标曲线
# - where: 只填充满足该条件的区域（这里是 x2 >= x1）
# - interpolate=True: 在交叉点处插值，确保填充区域连续
# - facecolor: 填充颜色
# - alpha: 透明度
ax.fill_betweenx(y, x1, x2, where=x2 >= x1, interpolate=True, facecolor="cornflowerblue", alpha=0.7)

# 在 x1 > x2 的区域填充红色
ax.fill_betweenx(y, x1, x2, where=x2 <= x1, interpolate=True, facecolor="darkred", alpha=0.7)

# 设置横轴的显示范围为 [-1.2, 1.2]
ax.set_xlim(-1.2, 1.2)

# 设置纵轴的显示范围为 [0, 2]
ax.set_ylim(0, 2)

# 添加网格线
# - ls=":" 设置线型为虚线
# - lw=1 设置线宽
# - color="gray" 设置颜色
# - alpha=0.5 设置透明度
ax.grid(ls=":", lw=1, color="gray", alpha=0.5)

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建一个包含 1 行 2 列子图的画布
fig, ax = plt.subplots(1, 2, figsize=(12, 5))  # 可指定宽高

# ---------------- 子图1：y = f(x) 的情况 ----------------
# 定义 x 范围
x = np.linspace(0, 2, 500)

# 曲线 y1 = sin(2πx)，y2 = 1.2*sin(3πx)
y1 = np.sin(2 * np.pi * x)
y2_raw = 1.2 * np.sin(3 * np.pi * x)

# 将 y2 中超过 1.0 的值设为 NaN，避免后续绘图出错
y2 = np.where(y2_raw > 1.0, np.nan, y2_raw)

# 绘制两条曲线
ax[0].plot(x, y1, color="k", lw=1, ls="-", label="y1")
ax[0].plot(x, y2, color="k", lw=1, ls="-", label="y2")

# 构造掩码，排除 y2 中的 NaN 值
valid = ~np.isnan(y2)

# 条件1：y2 >= y1 的区域
cond1 = np.logical_and(valid, y2 >= y1)
# 条件2：y2 <= y1 的区域
cond2 = np.logical_and(valid, y2 <= y1)

# 填充 y1 到 y2 之间的区域（蓝色部分：y2 >= y1）
ax[0].fill_between(x, y1, y2, where=cond1, interpolate=True, facecolor="cornflowerblue", alpha=0.7)

# 填充 y1 到 y2 之间的区域（红色部分：y2 <= y1）
ax[0].fill_between(x, y1, y2, where=cond2, interpolate=True, facecolor="darkred", alpha=0.7)

# 设置坐标轴范围
ax[0].set_xlim(0, 2)
ax[0].set_ylim(-1.2, 1.2)

# 添加网格线
ax[0].grid(ls=":", lw=1, color="gray", alpha=0.5)

# 设置标题
ax[0].set_title("Plot 1: y vs x")



# ---------------- 子图2：x = f(y) 的情况 ----------------
y = np.linspace(0, 2, 500)
x1 = np.sin(2 * np.pi * y)
x2_raw = np.sin(3 * np.pi * y)

# 阈值遮蔽：将超过 1.0 的值设为 NaN
x2 = np.where(x2_raw > 1.0, np.nan, x2_raw)

# 绘制两条曲线
ax[1].plot(x1, y, color="k", lw=1, ls="-", label="x1")
ax[1].plot(x2, y, color="k", lw=1, ls="-", label="x2")

# 有效区域：x1 和 x2 都不是 NaN
valid = ~np.isnan(x2) & ~np.isnan(x1)

# 条件区域：x2 >= x1 且值有效
cond1 = np.logical_and(valid, x2 >= x1)
# 条件区域：x2 <= x1 且值有效
cond2 = np.logical_and(valid, x2 <= x1)

# 填充区域
ax[1].fill_betweenx(y, x1, x2, where=cond1, interpolate=True, facecolor="cornflowerblue", alpha=0.7)
ax[1].fill_betweenx(y, x1, x2, where=cond2, interpolate=True, facecolor="darkred", alpha=0.7)

# 设置坐标轴范围和网格
ax[1].set_xlim(-1.2, 1.2)
ax[1].set_ylim(0, 2)
ax[1].grid(ls=":", lw=1, color="gray", alpha=0.5)
ax[1].set_title("Plot 2: x vs y")

```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# 初始化频率
f1_init = 2
f2_init = 3

# 创建画布和子图
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
plt.subplots_adjust(bottom=0.25)  # 给滑块留空间

# 设置 x 和 y 范围
x = np.linspace(0, 2, 500)
y = np.linspace(0, 2, 500)

# 初始化绘图函数
def update(f1, f2):
    # 清空子图
    for a in ax:
        a.clear()

    # ----------- 子图1：y = sin(fπx) ----------
    y1 = np.sin(f1 * np.pi * x)
    y2_raw = 1.2 * np.sin(f2 * np.pi * x)
    y2 = np.where(y2_raw > 1.0, np.nan, y2_raw)
    valid = ~np.isnan(y2) & ~np.isnan(y1)
    cond1 = np.logical_and(valid, y2 >= y1)
    cond2 = np.logical_and(valid, y2 <= y1)

    ax[0].plot(x, y1, color="k", lw=1)
    ax[0].plot(x, y2, color="k", lw=1)
    ax[0].fill_between(x, y1, y2, where=cond1, interpolate=True, facecolor="cornflowerblue", alpha=0.7)
    ax[0].fill_between(x, y1, y2, where=cond2, interpolate=True, facecolor="darkred", alpha=0.7)
    ax[0].set_xlim(0, 2)
    ax[0].set_ylim(-1.2, 1.2)
    ax[0].grid(ls=":", lw=1, color="gray", alpha=0.5)
    ax[0].set_title(f"Plot 1: y = sin({f1}πx), y2 = 1.2*sin({f2}πx)")

    # ----------- 子图2：x = sin(fπy) ----------
    x1 = np.sin(f1 * np.pi * y)
    x2_raw = np.sin(f2 * np.pi * y)
    x2 = np.where(x2_raw > 1.0, np.nan, x2_raw)
    valid = ~np.isnan(x2) & ~np.isnan(x1)
    cond1 = np.logical_and(valid, x2 >= x1)
    cond2 = np.logical_and(valid, x2 <= x1)

    ax[1].plot(x1, y, color="k", lw=1)
    ax[1].plot(x2, y, color="k", lw=1)
    ax[1].fill_betweenx(y, x1, x2, where=cond1, interpolate=True, facecolor="cornflowerblue", alpha=0.7)
    ax[1].fill_betweenx(y, x1, x2, where=cond2, interpolate=True, facecolor="darkred", alpha=0.7)
    ax[1].set_xlim(-1.2, 1.2)
    ax[1].set_ylim(0, 2)
    ax[1].grid(ls=":", lw=1, color="gray", alpha=0.5)
    ax[1].set_title(f"Plot 2: x = sin({f1}πy), x2 = sin({f2}πy)")

    plt.draw()

# 初次绘图
update(f1_init, f2_init)

# 添加滑块控件
ax_f1 = plt.axes([0.2, 0.12, 0.65, 0.03])
ax_f2 = plt.axes([0.2, 0.06, 0.65, 0.03])

slider_f1 = Slider(ax_f1, 'f1', 0.5, 5.0, valinit=f1_init, valstep=0.1)
slider_f2 = Slider(ax_f2, 'f2', 0.5, 5.0, valinit=f2_init, valstep=0.1)

# 当滑块变化时，更新图像
def on_slider_change(val):
    f1 = slider_f1.val
    f2 = slider_f2.val
    update(f1, f2)

slider_f1.on_changed(on_slider_change)
slider_f2.on_changed(on_slider_change)

plt.show()
```

### 椭圆

```python
import matplotlib.pyplot as plt          # 引入 matplotlib 的绘图模块
import numpy as np                       # 引入 numpy 用于数值计算
from matplotlib.patches import Ellipse   # 从 patches 模块引入 Ellipse 类，用于绘制椭圆

# 创建一个包含两个子图的图形对象，每个子图纵横比为 1:1（等比例）
fig, ax = plt.subplots(1, 2, subplot_kw={"aspect": "equal"})

# ---------------------- 子图 1：不同角度的椭圆 ----------------------

# 创建一个角度数组，linspace(start, end, num)：在0~135度之间均匀取4个角度
angles = np.linspace(0, 135, 4)

# 使用列表推导式，创建4个椭圆，中心在 (2,2)，长轴4，短轴2，角度分别是 angles 中的值
ellipses = [Ellipse((2, 2), width=4, height=2, angle=a) for a in angles]

# 遍历所有椭圆对象，将它们添加到第一个子图中
for e in ellipses:
    ax[0].add_patch(e)            # 将椭圆添加到子图 ax[0]
    e.set_alpha(0.4)              # 设置透明度为 0.4
    e.set_color("cornflowerblue")# 设置颜色为浅蓝色

# 设置第一个子图的显示范围
ax[0].axis([-1, 5, -1, 5])        # x 轴 [-1, 5]，y 轴 [-1, 5]

# ---------------------- 子图 2：100 个随机椭圆 ----------------------

# 创建一个从 0 到 99 的整数数组，代表 100 个椭圆
num = np.arange(0, 100, 1)

# 使用列表推导式生成 100 个随机椭圆：
#   - 随机中心位置：np.random.rand(2) * 10，范围 [0,10)
#   - 宽度、高度、角度分别是 0~1 范围内的随机数
ellipses = [
    Ellipse(
        xy=np.random.rand(2) * 10,         # 随机中心点坐标 (x, y)
        width=np.random.rand(1)[0],        # 随机宽度（长轴）
        height=np.random.rand(1)[0],       # 随机高度（短轴）
        angle=np.random.rand(1)[0] * 360   # 随机旋转角度（0~360度）
    )
    for _ in num
]

# 遍历所有椭圆，设置样式并添加到第二个子图
for e in ellipses:
    ax[1].add_patch(e)               # 添加到子图 ax[1]
    e.set_alpha(np.random.rand(1)[0])# 设置 0~1 的随机透明度
    e.set_color(np.random.rand(3))   # 设置随机 RGB 颜色（3个浮点数）

# 设置第二个子图的显示范围
ax[1].axis([-1, 11, -1, 11])         # x 轴 [-1, 11]，y 轴 [-1, 11]

# 自动调整子图之间的布局，避免重叠
plt.tight_layout()

# 显示图形窗口
plt.show()
```

### 矩形

```python
import matplotlib.pyplot as plt              # 引入 matplotlib 绘图模块
import numpy as np                           # 引入 numpy 数值计算模块
from matplotlib.patches import Rectangle     # 从 matplotlib.patches 引入矩形类

# 创建一个图形和坐标轴对象，设置子图纵横比为1（x、y等单位缩放）
fig, ax = plt.subplots(subplot_kw={"aspect": "equal"})

# 创建两个 x 坐标数组，用于绘制屋顶填充区域
x1 = np.arange(1, 2.6, 0.1)           # 从 1 到 2.5，步长 0.1，用于左半边屋顶
y1 = x1 + 2                           # 对应的 y 值，形成斜线 y = x + 2

x2 = np.arange(2.5, 4.1, 0.1)         # 从 2.5 到 4，用于右半边屋顶
y2 = -x2 + 7                          # 对应的 y 值，形成斜线 y = -x + 7

# ---------------------- 设置背景颜色 ----------------------
rectangle = ax.patch                 # 获取坐标轴背景
rectangle.set_facecolor("lightskyblue")  # 设置背景颜色为浅天蓝色

# ---------------------- 房屋主体元素 ----------------------

# 房屋主体：左下角 (1,0)，宽 3，高 3，白底红边
rectangle1 = Rectangle((1, 0), 3, 3, facecolor="w", edgecolor="rosybrown")

# 门：左下角 (1.5, 0)，宽 1，高 1.5，带有竖线阴影 |||
rectangle2 = Rectangle((1.5, 0), 1, 1.5, facecolor="w", edgecolor="rosybrown", hatch="|||")

# 窗户：左下角 (2.9, 1.7)，宽 0.6，高 0.6
rectangle3 = Rectangle((2.9, 1.7), 0.6, 0.6, facecolor="w", edgecolor="rosybrown")

# 将房屋元素打包成列表，后面统一绘制
rectangle_list = [rectangle1, rectangle2, rectangle3]

# ---------------------- 屋顶线条（轮廓） ----------------------

# 屋顶轮廓线，三点组成：左墙顶 -> 屋顶顶点 -> 右墙顶
ax.plot([1, 2.5, 4], [3, 4.5, 3], color="rosybrown")

# 窗户竖线条（窗框）
ax.plot([3.2, 3.2], [1.7, 2.3], color="rosybrown")

# 窗户横线条（窗框）
ax.plot([2.9, 3.5], [2.0, 2.0], color="rosybrown")

# ---------------------- 屋顶填充 ----------------------

# 使用 fill_between 填充左半边屋顶区域
ax.fill_between(x1, 3, y1, color="w", interpolate=True)

# 使用 fill_between 填充右半边屋顶区域
ax.fill_between(x2, 3, y2, color="w", interpolate=True)

# ---------------------- 添加矩形元素 ----------------------

# 遍历矩形列表，逐个添加到坐标轴上
for rect in rectangle_list:
    ax.add_patch(rect)

# ---------------------- 设置显示范围 ----------------------

ax.axis([0, 5, 0, 6])  # 设置 x 轴 [0,5]，y 轴 [0,6] 的显示范围

# ---------------------- 显示图形 ----------------------

plt.show()
```

### 判别分析示意图

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建图像和坐标轴
fig, ax = plt.subplots()

# 样本数量
num = 50

# 生成 50 个二维随机样本点，每个分量取值在 [0, 10)
sample = 10 * np.random.rand(num, 2)
var1 = sample[:, 0]  # 第一维特征
var2 = sample[:, 1]  # 第二维特征

# 设置分类判别的阈值线
td = 12
df = 2 * var1 + var2  # 判别函数（例如感知器的一种线性组合）

# 使用掩码将数据按分类分成两组（小于和大于等于阈值）
# 类别A：df < td（线下区域）
cates11 = np.ma.masked_where(df >= td, var1)
cates12 = np.ma.masked_where(df >= td, var2)

# 类别B：df >= td（线上或线以上）
cates21 = np.ma.masked_where(df < td, var1)
cates22 = np.ma.masked_where(df < td, var2)

# 绘制类别A的点（方形、颜色和大小与var1相关）
scatter_a = ax.scatter(var1, var2, 
                       s=cates11 * 50, 
                       marker="s", 
                       c=cates11, 
                       cmap="viridis", 
                       label="Class A (Below line)")

# 绘制类别B的点（圆形、颜色和大小与var1相关）
scatter_b = ax.scatter(var1, var2, 
                       s=cates21 * 50, 
                       marker="o", 
                       c=cates21, 
                       cmap="plasma", 
                       label="Class B (Above line)")

# 绘制决策边界线 y = -2x + 12，对应判别函数 df = 12
x_line = np.linspace(-1, 11, 100)
y_line = -2 * x_line + 12
ax.plot(x_line, y_line, color="blue", lw=2, alpha=0.6, label="Decision Boundary (2x + y = 12)")

# 设置坐标轴范围
ax.axis([-1, 11, -1, 11])

# 添加图例
ax.legend()

# 添加颜色条（针对 var1 的值，解释颜色意义）
cbar = plt.colorbar(scatter_a, ax=ax, orientation='vertical', pad=0.02)
cbar.set_label('var1 Value (used for size & color)')

# 设置标题和标签
ax.set_title("Linear Classification Visualization")
ax.set_xlabel("Feature var1")
ax.set_ylabel("Feature var2")

# 显示图像
plt.tight_layout()
plt.show()
```

### 直方图中添加概率密度曲线

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import numpy as np

# ------------------ 图像全局设置 ------------------

# 设置中文字体为仿宋，避免中文乱码
mpl.rcParams["font.sans-serif"] = ["FangSong"]
# 显示负号
mpl.rcParams["axes.unicode_minus"] = False

# ------------------ 数据准备 ------------------

# 正态分布参数：均值 μ = 60，标准差 σ = 2
mu = 60.0
sigma = 2.0

# 生成500个正态分布随机样本（N(60, 2²)）
x = mu + sigma * np.random.randn(500)

# 设置直方图的分组数
bins = 50

# 生成用于绘制正态曲线和平滑积分区域的 x 值
x_curve = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y_curve = (1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * ((x_curve - mu) / sigma) ** 2)

# ------------------ 构造积分阴影区域 ------------------

# 积分范围 [μ - 2σ, μ + 2σ]
integ_x = np.linspace(mu - 2*sigma, mu + 2*sigma, 1000)
integ_y = (1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * ((integ_x - mu) / sigma) ** 2)

# 构造多边形区域（闭合）
area = [(mu - 2*sigma, 0), *zip(integ_x, integ_y), (mu + 2*sigma, 0)]

# ------------------ 图像绘制 ------------------

fig, ax = plt.subplots(1, 1)

# 绘制归一化直方图
n, bins, patches = ax.hist(
    x,
    bins=bins,
    density=True,              # 密度归一化
    histtype="bar",            # 柱状样式
    facecolor="cornflowerblue",
    edgecolor="white",
    alpha=0.75
)

# 绘制正态分布密度曲线（平滑）
ax.plot(x_curve, y_curve, color="orange", linestyle="--", linewidth=2, label="正态分布曲线")

# 添加积分区域阴影（Polygon 多边形）
poly = Polygon(area, facecolor="gray", edgecolor="k", alpha=0.6, closed=True)
ax.add_patch(poly)

# 添加积分公式标签，使用相对坐标放置（ax.transAxes）
plt.text(0.45, 0.2, r"$\int_{\mu-2\sigma}^{\mu+2\sigma} y\,\mathrm{d}x$", fontsize=20, transform=ax.transAxes)

# 添加公式说明文本
ax.text(
    54, 0.2,
    r"$y=\frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$",
    color="red", fontsize=20
)

# 设置坐标轴标签和标题
ax.set_xlabel("weight")                   # X轴：体重
ax.set_ylabel("概率密度")                 # Y轴：概率密度
ax.set_title(r"体重的直方图: $\mu=60.0$, $\sigma=2.0$", fontsize=16)

# 添加网格线
ax.grid(linestyle=":", linewidth=1, color="gray", alpha=0.2)

# 自动调整子图，防止标签遮挡
plt.tight_layout()

# 显示图像
plt.show()
```

### 绘图区嵌套子绘图区

```python
import matplotlib.pyplot as plt
import numpy as np

# ------------------- 参数设定 -------------------
mu = 75.0                 # 正态分布均值 μ（即期望值）
sigma = 15.0              # 正态分布标准差 σ，控制数据分散程度
bins = 20                 # 直方图的柱子数量（分箱数）
x = np.linspace(1, 100, 200)          # 创建一个包含200个点的线性空间，从1到100，作为X轴的数据
y = np.random.normal(mu, sigma, 200)  # 生成200个正态分布数据，均值为mu，标准差为sigma

# ------------------- 主图 -------------------
fig, ax = plt.subplots(figsize=(8, 6))  # 创建主图的画布和子图
# fig: 图像对象，ax: 坐标轴对象
# figsize=(8,6) 是设置画布的宽度为8英寸，高度为6英寸

ax.plot(x, y, ls="-", lw=2, color="steelblue")  # 绘制折线图
# x: 横坐标数据
# y: 纵坐标数据
# ls="-": 线型设置为实线
# lw=2: 线宽为2
# color="steelblue": 线的颜色为“钢蓝色”

ax.set_ylim(10, 170)  # 设置 y 轴范围，防止数据超出显示范围

# ------------------- 左上角小直方图 -------------------
# 创建一个嵌套子图（即右侧小区域）
# [left, bottom, width, height] 表示该小图的位置和大小（这些值是相对于主图大小的比例）
inset1 = fig.add_axes([0.2, 0.6, 0.2, 0.2], facecolor="k")  
# 参数：[左，底部，宽度，高度]，值的范围是 [0, 1]，0表示边界，1表示最大边界

# 绘制直方图
# y: 数据
# bins: 分组数（即柱子的数量）
# color: 填充颜色为“cornflowerblue”
count, bin_edges, patches = inset1.hist(y, bins=bins, color="cornflowerblue")
# count: 每个柱子的高度（即每个区间的频数）
# bin_edges: 每个柱子的边界
# patches: 绘制的柱状图对象

# 设置 y 轴范围，显示频数的区间
inset1.set_ylim(0, 28)
inset1.set_xticks([])  # 不显示 x 轴刻度
inset1.set_yticks([])  # 不显示 y 轴刻度

# ------------------- 左上角插入正态分布密度曲线 -------------------
# 在上一个小图的右上角插入一条正态分布的概率密度曲线
inset1_curve = fig.add_axes([0.21, 0.72, 0.05, 0.05])  # 小图的位置和大小
# 计算正态分布密度函数（pdf）
y1 = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((bin_edges - mu) ** 2) / (2 * sigma ** 2))
# y1: 给定 bin_edges 上正态分布的概率密度

# 绘制正态分布密度曲线
# ls="-": 实线
# color="r": 线的颜色为红色
inset1_curve.plot(bin_edges, y1, ls="-", color="r")

# 设置坐标轴刻度为空，确保小图没有显示坐标轴
inset1_curve.set_xticks([])
inset1_curve.set_yticks([])

# ------------------- 右上角累计直方图 -------------------
# 创建另一个嵌套子图，用来绘制累计直方图
inset2 = fig.add_axes([0.65, 0.6, 0.2, 0.2], facecolor="k")  
# 参数：[左，底部，宽度，高度]，该图放在主图右上角

# 绘制累计直方图
# y: 数据
# bins: 分组数
# color: 填充颜色为“cornflowerblue”
# density=True: 归一化，确保直方图的面积为1
# cumulative=True: 累计直方图，将每个柱子显示为到该点的累计值
# histtype="step": 只绘制直方图的轮廓线（无填充）
count, bin_edges, patches = inset2.hist(
    y, bins=bins, color="cornflowerblue",
    density=True, cumulative=True, histtype="step"
)

# 设置 y 轴范围，显示累计值的区间
inset2.set_ylim(0, 1.0)
inset2.set_xticks([])  # 不显示 x 轴刻度
inset2.set_yticks([])  # 不显示 y 轴刻度

# ------------------- 右上角插入累计正态分布曲线 -------------------
# 在右上角插入累计正态分布的曲线
inset2_curve = fig.add_axes([0.66, 0.72, 0.05, 0.055])  # 小图的位置和大小
# 计算正态分布的累计密度（累积分布函数）
y2 = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((bin_edges - mu) ** 2) / (2 * sigma ** 2))
y2 = y2.cumsum()     # 对 y2 执行累加操作
y2 = y2 / y2[-1]     # 将累计值归一化为 [0, 1]，最后一个值为 1

# 绘制累计分布曲线
# ls="-": 实线
# color="r": 线的颜色为红色
inset2_curve.plot(bin_edges, y2, ls="-", color="r")

# 设置坐标轴刻度为空，确保小图没有显示坐标轴
inset2_curve.set_xticks([])
inset2_curve.set_yticks([])

# ------------------- 显示图形 -------------------
plt.show()  # 展示绘制的图形
```

### 设置一般化的日期刻度线

```python
import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np

# ------------------- 创建图表 -------------------
fig, ax = plt.subplots()  # 创建图表和坐标轴

# ------------------- 日期规则和定位器 -------------------
# mdates.rrulewrapper 用来生成特定规则的日期规则包装器
# 规则：每年 (YEARLY)，从开始日期起每隔2年，计算范围内的所有符合该规则的日期
rule = mdates.rrulewrapper(mdates.YEARLY, byeaster=0, interval=2)
# byeaster=0 表示忽略复活节假期
# interval=2 表示每两年进行一次

# 使用 RRuleLocator 来将规则应用于定位器，以确定主要刻度的间隔
loc = mdates.RRuleLocator(rule)

# ------------------- 日期格式化 -------------------
# 设置 x 轴的日期格式
dateFmt = mdates.DateFormatter("%m/%d/%y")
ax.xaxis.set_major_locator(loc)  # 设置 x 轴的主要定位器为 RRuleLocator
ax.xaxis.set_major_formatter(dateFmt)  # 设置 x 轴的日期格式

# ------------------- 设置图表的外观 -------------------
# tick_params 用于调整坐标轴刻度的外观
# direction="out" 将刻度线朝外显示
# labelsize=10 设置刻度标签的字体大小为 10
ax.tick_params(axis="both", direction="out", labelsize=10)

# ------------------- 日期范围和数据生成 -------------------
# 定义日期范围，从2004年5月17日到2016年6月4日
date1 = datetime.date(2004, 5, 17)
date2 = datetime.date(2016, 6, 4)

# 设置日期间隔，每次增加5天
delta = datetime.timedelta(days=5)

# 生成日期列表
dates = mdates.drange(date1, date2, delta)

# 生成与日期对应的随机数据
# 这里的随机数据是根据正态分布生成的，均值为120，标准差为12
y = np.random.normal(120, 12, len(dates))

# ------------------- 使用 plot 绘制数据 -------------------
# 使用 ax.plot 绘制日期与数据的关系
# 将日期转化为 `matplotlib` 可识别的格式，传递给 plot() 方法
ax.plot(mdates.num2date(dates), y, "b-", alpha=0.7)

# 自动旋转 x 轴日期标签，避免重叠
fig.autofmt_xdate()

# ------------------- 显示图表 -------------------
plt.show()
```

### 文本框的样式

```python
import matplotlib.patches as patches
import matplotlib.pyplot as plt

# 创建图像窗口
# - figsize=(8, 9): 图像大小（单位为英寸）
# - dpi=72: 每英寸点数（像素密度）
fig = plt.figure(1, figsize=(10, 10), dpi=72)

# 设置字体大小（基于dpi计算）
fontsize = 16  # 可改为固定值如 fontsize=12 以便清晰显示

# 添加坐标轴子图（1行1列的第1个图）
# - frameon=False: 不显示坐标轴边框
# - xticks=[], yticks=[]: 不显示刻度
ax = fig.add_subplot(1, 1, 1, frameon=False, xticks=[], yticks=[])

# 获取所有可用的 BoxStyle 样式
boxStyles = patches.BoxStyle.get_styles()
boxStyleNames = list(boxStyles.keys())
boxStyleNames.sort()  # 排序方便查看

# 遍历所有样式并在图中显示
for i, name in enumerate(boxStyleNames):
    # ax.text(...) 添加文本框
    # - x, y: 文本框的位置（使用 Axes 的坐标系统）
    # - name: 显示的样式名称
    # - ha="center": 文本水平居中
    # - transform=ax.transAxes: 使用坐标轴的相对坐标 (0~1)
    # - bbox: 文本框样式设置（boxstyle为当前样式，白色填充、黑边框）
    ax.text(
        float(i + 0.5) / len(boxStyleNames),             # x位置：平均分布
        (float(len(boxStyleNames)) - 0.5 - i) / len(boxStyleNames),  # y位置：从上到下
        name,                                            # 显示样式名
        ha="center",                                     # 水平居中对齐
        size=fontsize,                                   # 字体大小
        transform=ax.transAxes,                          # 使用轴坐标系
        bbox=dict(boxstyle=name, fc="w", ec="k")         # 文本框样式
    )

# 显示图形（在循环外）
plt.show()
```

### 文本注释箭头的样式

```python
import matplotlib.patches as patches
import matplotlib.pyplot as plt

# 创建图像窗口，大小为10x10英寸，分辨率为72 DPI
fig = plt.figure(1, figsize=(8, 8), dpi=72)

# 设置字体大小（根据 DPI 缩放）
fontsize = 0.2 * fig.dpi

# 创建无边框坐标轴
ax = fig.add_subplot(1, 1, 1, frameon=False)

# 获取并排序所有箭头样式名称
arrowStyles = patches.ArrowStyle.get_styles()
arrowStyleNames = sorted(arrowStyles.keys())

# 设置坐标范围
ax.set_xlim(0, len(arrowStyleNames) + 2.5)
ax.set_ylim(-2, len(arrowStyleNames))

# 遍历所有箭头样式
for i, name in enumerate(arrowStyleNames):
    start_x = float(len(arrowStyleNames)) + 1.2 - i
    start_y = float(len(arrowStyleNames)) - 2.8 - i

    # 添加一个圆形作为起点
    p = patches.Circle((start_x, start_y), radius=0.2, color="steelblue", alpha=1.0)
    ax.add_patch(p)

    # 添加注释，箭头连接圆形和文字
    ax.annotate(
        text=name,  # 显示箭头样式名
        xy=(start_x, start_y),  # 箭头起点（圆心）
        xytext=(start_x - 2.2, start_y - 0.2),  # 文本位置
        xycoords='data',
        ha="center",
        size=fontsize,
        bbox=dict(              # ✅ 注意：这是文本框样式，必须放在这里！
            boxstyle="round",  # 可选 box 样式，比如 round, square
            fc="w",            # 背景色 white
            ec="k"             # 边框色 black
        ),
        arrowprops=dict(       # 箭头参数设置
            arrowstyle=name,   # 使用当前样式
            fc="k",            # 填充颜色
            ec="k",            # 边框颜色
            patchB=p,          # 起点 patch（圆形）
            shrinkA=5,         # 起点端缩短
            shrinkB=5,         # 终点端缩短
            connectionstyle="arc3,rad=0.3"  # 曲线样式
        )
    )

# 关闭坐标轴显示
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# -------------------- 数据准备 --------------------

# 生成从 0 到 3π 的 500 个点，作为弧度值
x_rad = np.linspace(0, 3 * np.pi, 500)

# 对应的角度值（将弧度转换为角度）
x_deg = np.degrees(x_rad)

# 计算余弦值
y_cos = np.cos(x_rad)

# -------------------- 创建图像 --------------------

fig, ax = plt.subplots(2, 1, figsize=(8, 6))

# 第一个子图：弧度为横轴
ax[0].plot(x_rad, y_cos, linestyle='-', linewidth=2, color='black')
ax[0].set_xlabel("x (radians)")
ax[0].set_ylabel("cos(x)")
ax[0].set_title("Cosine Function (radians)")
ax[0].grid(True)

# 第二个子图：角度为横轴（用 x_deg 作为横坐标）
ax[1].plot(x_deg, y_cos, linestyle='--', linewidth=2, color='cornflowerblue')
ax[1].set_xlabel("x (degrees)")
ax[1].set_ylabel("cos(x)")
ax[1].set_title("Cosine Function (degrees)")
ax[1].grid(True)

# 调整子图间距
fig.subplots_adjust(hspace=0.3)

# 显示图形
plt.show()
```

### 厘米和英寸的实现

```python
import matplotlib.pyplot as plt
import numpy as np

# 单位转换：1 英寸 = 2.54 厘米
cm_to_inch = 1 / 2.54

# 原始数据（以厘米为单位）
x_cm = np.linspace(0, 10, 6)

# 手动转换
x_inch = x_cm * cm_to_inch

# 创建图形和子图
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

# --------------------- 第一张图：cm vs cm ---------------------
ax[0, 0].plot(x_cm, x_cm, ls='-', lw=3, c='k', marker='s')
ax[0, 0].set_title("cm vs cm")
ax[0, 0].set_xlabel("x (cm)")
ax[0, 0].set_ylabel("y (cm)")
ax[0, 0].grid(True)

# --------------------- 第二张图：cm vs inch ---------------------
ax[0, 1].plot(x_cm, x_cm * cm_to_inch, ls='--', lw=3, c='cornflowerblue', marker='s')
ax[0, 1].set_title("cm vs inch")
ax[0, 1].set_xlabel("x (cm)")
ax[0, 1].set_ylabel("y (inch)")
ax[0, 1].set_xlim(2, 8)
ax[0, 1].grid(True)

# --------------------- 第三张图：inch vs cm ---------------------
ax[1, 0].plot(x_cm * cm_to_inch, x_cm, ls='-.', lw=3, c='gold', marker='s')
ax[1, 0].set_title("inch vs cm")
ax[1, 0].set_xlabel("x (inch)")
ax[1, 0].set_ylabel("y (cm)")
ax[1, 0].set_xlim(2 * cm_to_inch, 8 * cm_to_inch)
ax[1, 0].grid(True)

# --------------------- 第四张图：inch vs inch ---------------------
ax[1, 1].plot(x_inch, x_inch, ls=':', lw=3, c='purple', marker='s')
ax[1, 1].set_title("inch vs inch")
ax[1, 1].set_xlabel("x (inch)")
ax[1, 1].set_ylabel("y (inch)")
ax[1, 1].grid(True)

# 自动布局
plt.tight_layout()
plt.show()
```
